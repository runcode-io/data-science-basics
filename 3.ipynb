# Statistics and probability

## 3.1 Statistics

<b>Statistics</b> is the study of collecting, organizing, analyzing, interpreting, and presenting data. It involves using statistical techniques to summarize and describe data, and to make predictions or inferences about a population based on a sample of data.

There are many statistical techniques that are commonly used in data science, and the specific technique that is used can depend on the nature of the data and the goals of the analysis. Here are some examples of statistical techniques that are commonly used in data science, along with a brief description of each technique and sample Python code that illustrates how to use the technique:

### 1. Mean:

The mean is the average value of a set of numbers. To compute the mean of a set of numbers in Python, you can use the mean() function from the statistics module.

```python
from statistics import mean

# Compute the mean of a list of numbers
numbers = [1, 2, 3, 4, 5]
mean_val = mean(numbers)
print(mean_val)
```

### 2. Median:

The median is the middle value in a set of numbers. To compute the median of a set of numbers in Python, you can use the median() function from the statistics module.

```python
from statistics import median

# Compute the median of a list of numbers
numbers = [1, 2, 3, 4, 5]
median_val = median(numbers)
print(median_val)
```

### 3. Mode:

The mode is the most frequently occurring value in a set of numbers. To compute the mode of a set of numbers in Python, you can use the mode() function from the statistics module.

```python
from statistics import mode

# Compute the mode of a list of numbers
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
mode_val = mode(numbers)
print(mode_val)
```

### 4. Standard Deviation: 

The standard deviation is a measure of the dispersion of a set of numbers. To compute the standard deviation of a set of numbers in Python, you can use the stdev() function from the statistics module.

```python
from statistics import stdev

# Compute the standard deviation of a list of numbers
numbers = [1, 2, 3, 4, 5]
stdev_val = stdev(numbers)
print(stdev_val)
```

### 5. Correlation:

Correlation is a measure of the relationship between two sets of numbers. To compute the correlation between two sets of numbers in Python, you can use the pearsonr() function from the scipy.stats module.

```python
from scipy.stats import pearsonr

# Compute the Pearson correlation coefficient between two sets of numbers
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
corr, p_value = pearsonr(x, y)
print(corr)
```

## 3.2 Probability

<b>Probability</b> is the study of random events and the likelihood of their occurrence. It is used to describe the likelihood of certain outcomes in statistical experiments. It is a mathematical concept that is used to describe and analyze random phenomena, such as the roll of a dice or the flip of a coin.

Here are some important probability concepts and techniques, along with examples of how to implement them in Python:

1. <b>Probability mass function (PMF):</b> This is a function that assigns a probability to each possible value of a discrete random variable. For example, the PMF of a dice roll can be represented as a table or a plot that shows the probability of rolling each number from 1 to 6. In Python, you can use the scipy.stats library to create a PMF and plot it.

    Example:

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Create a PMF for a dice roll
n = 1  # number of dice
k = 6  # number of sides
p = 1/k  # probability of rolling each side
pmf = binom.pmf(range(1, k+1), n, p)

# Plot the PMF
plt.bar(range(1, k+1), pmf)
plt.xlabel("Number")
plt.ylabel("Probability")
plt.show()
```

2. <b>Probability density function (PDF):</b> This is a function that describes the continuous probability distribution of a random variable. For example, the PDF of a normal distribution can be represented as a bell-shaped curve that shows the probability of observing different values within a certain range. In Python, you can use the scipy.stats library to create a PDF and plot it.

    Example:

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Create a PDF for a normal distribution
mean = 0
std = 1
pdf = norm.pdf(np.linspace(-5, 5, 100), mean, std)

# Plot the PDF
plt.plot(np.linspace(-5, 5, 100), pdf)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

3. <b>Cumulative distribution function (CDF):</b> This is a function that describes the probability that a random variable is less than or equal to a certain value. For example, the CDF of a normal distribution can be represented as a curve that shows the probability of observing a value less than or equal to a certain number. In Python, you can use the scipy.stats library to create a CDF and plot it.

    Example:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# Create a CDF for a normal distribution
mean = 0
std = 1
cdf = norm.cdf(np.linspace(-5, 5, 100), mean, std)

# Plot the CDF
plt.plot(np.linspace(-5, 5, 100), cdf)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

4. <b>Conditional probability</b> is the probability of an event occurring, given that another event has occurred. For example, you might want to know the probability of rolling a 6 on a dice, given that the previous roll was a 4. Here is an example of how to calculate and plot conditional probability in Python:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a contingency table that shows the number of times each pair of events occurred
cont_table = np.array([[5, 1],
                       [1, 5]])

# Calculate the conditional probability of rolling a 6, given that the previous roll was a 4
p_6_given_4 = cont_table[1, 1] / cont_table[1, :].sum()
print(f"Probability of rolling a 6, given that the previous roll was a 4: {p_6_given_4:.2f}")

# Calculate the conditional probability of rolling a 4, given that the previous roll was a 6
p_4_given_6 = cont_table[0, 1] / cont_table[0, :].sum()
print(f"Probability of rolling a 4, given that the previous roll was a 6: {p_4_given_6:.2f}")

# Plot the conditional probability of rolling a 6, given the previous roll
labels = ["4", "6"]
p_given_4 = [1-p_6_given_4, p_6_given_4]
p_given_6 = [1-p_4_given_6, p_4_given_6]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, p_given_4, width, label="Given 4")
rects2 = ax.bar(x + width/2, p_given_6, width, label="Given 6")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Probability")
ax.set_title("Conditional probability of rolling a 6")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
