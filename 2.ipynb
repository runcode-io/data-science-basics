# Data Visualization

Data visualization is an important part of data science, as it allows you to explore and understand your data, and communicate your findings to others. There are many tools and libraries available for creating data visualizations, and two of the most popular ones are Matplotlib and Plotly.

## 2.1 Matplotlib

<b>Matplotlib</b> is a popular library for creating static, animated, and interactive visualizations in Python. Here are a few examples of different types of visualizations that you can create using Matplotlib:

1.<b>Line plots:</b> Line plots are commonly used to visualize time series data.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Create the plot
plt.plot(x, y)
# Show the plot
plt.show()
```

2. <b>Scatter plots:</b> Scatter plots are used to visualize the relationship between two numeric variables.

```python
import numpy as np

# Create some sample data
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y)
plt.show()
```

3. <b>Bar plots:</b> Bar plots are used to visualize categorical data.

```python
# Create some sample data
categories = ['cat1', 'cat2', 'cat3']
values = [1, 2, 3]

# Create a bar plot
plt.bar(categories, values)
plt.show()
```

4. <b>Histograms:</b> Histograms are used to visualize the distribution of a numeric variable.

```python
# Create some sample data
data = np.random.normal(10, 5, 1000)

# Create a histogram
plt.hist(data, bins=20)
plt.show()
```

5. <b>Pie charts:</b> Pie charts are used to visualize the proportions of different categories.

```python
# Create some sample data
categories = ['cat1', 'cat2', 'cat3']
values = [1, 2, 3]

# Create a pie chart
plt.pie(values, labels=categories)
plt.show()
```

## 2.2 Seaborn

<b>Seaborn</b> is a library for Python that is built on top of Matplotlib and provides a higher-level interface for creating statistical graphics. Seaborn makes it easier to create many different types of plots and can be more visually appealing than Matplotlib. Here are a few examples of different types of visualizations that you can create using Seaborn:

```python
# pip install seaborn
```

```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

1. <b>Joint Plot:</b> shows the relationship between two numerical variables. It is equivalent to matplotlib's scatterplot.

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

2. <b>Distribution Plot:</b> shows distribution of a univariate set of observations. Other distribution plots are rugplot and kdeplot.

```python
sns.displot(tips['total_bill'], kde=False)
plt.show()
```

## 2.3 Plotly

<b>Plotly</b> is a library for creating interactive visualizations in Python. It allows you to create a wide range of static, animated, and interactive visualizations in a simple and intuitive way. Here are a few examples of different types of plots that you can create using Plotly:

1. <b>Line plots:</b> Line plots can be used to visualize time series data or the trend of a continuous variable over time.

```python
import plotly.express as px

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 2, 1]

# Create a line plot
fig = px.line(x=x, y=y)
fig.show()
```

2. <b>Scatter plots:</b> Scatter plots can be used to visualize the relationship between two numeric variables.

```python
import numpy as np

# Create some sample data
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
fig = px.scatter(x=x, y=y)
fig.show()
```

3. <b>Bar plots:</b> Bar plots can be used to visualize categorical data.

```python
# Create some sample data
categories = ['cat1', 'cat2', 'cat3']
values = [1, 2, 3]

# Create a bar plot
fig = px.bar(x=categories, y=values)
fig.show()
```

4. <b>Histograms:</b> Histograms can be used to visualize the distribution of a numeric variable.

```python
# Create some sample data
data = np.random.normal(10, 5, 1000)

# Create a histogram
fig = px.histogram(x=data, nbins=20)
fig.show()
```

5. <b>Pie charts</b>: Pie charts can be used to visualize the proportions of different categories.

```python
import plotly.express as px

# Create some sample data
categories = ['cat1', 'cat2', 'cat3']
values = [1, 2, 3]

# Create a pie chart
fig = px.pie(values=values, names=categories)
fig.show()
```
